```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data training

rho = 10
#rho_fungsi = np.exp(-alpha(x-c)^2)
epsilon = 1
delta_x = 0.5
jumlah_node = 100
iteration_num = 30000

A = np.zeros((jumlah_node))#guess
B = np.zeros((jumlah_node))#output

j = np.arange(0,jumlah_node)
for k in range(iteration_num):
    for i in range(1, jumlah_node-1):
        B[i] = (((rho/epsilon) * delta_x**2) + A[i+1] + A[i-1])/2
        A[i] = B[i]
print (B)

plt.plot(j, B)
plt.show()
```
