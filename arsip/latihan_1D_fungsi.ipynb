```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
alpha = 0.2
c = 2
epsilon = 1
delta_x = 0.5
jumlah_node = 100
iteration_num = 300

temp = np.zeros((jumlah_node))#guess
phi = np.zeros((jumlah_node))#output
j = np.arange(0,jumlah_node)
data_rho = []

def gaussian_distribution(konstanta_alpha, konstanta_c, m):
    return np.exp(-konstanta_alpha*(j[m]-konstanta_c)**2) #tinggal definisikan aja j[i] diganti apa

for k in range(iteration_num):
    for i in range(1, jumlah_node-1):
        rho = gaussian_distribution(alpha,c,i)
        data_rho.append(rho)
        phi[i] = ((data_rho[i-1]*delta_x**2) + temp[i+1] + temp[i-1])/2
        temp[i] = phi[i]

np.savetxt('data_1D_training_label.csv', phi, fmt = '%.8f', delimiter=',')
np.savetxt('data_1D_training_input.csv', data_rho, fmt = '%.8f', delimiter=',')
print (phi)

plt.plot(j, phi)
plt.show()
```
