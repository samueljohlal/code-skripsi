```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
epsilon = 1
delta_x = 0.5
jumlah_node = 100
iteration_num = 15000

temp = np.zeros((jumlah_node))#guess
phi = np.zeros((jumlah_node))#output
j = np.arange(0,jumlah_node)
data_rho = []
mu = 100
sigma = 50

#gaussian distribution
for r in range(jumlah_node):
    gauss = random.gauss(mu, sigma)
    data_rho.append(gauss)

#syarat batas
data_rho[0] = 0
data_rho[jumlah_node-1] = 0

#gauss-seidel method
for k in range(iteration_num):
    for i in range(1, jumlah_node-1):
        phi[i] = ((data_rho[i]*delta_x**2) + temp[i+1] + temp[i-1])/2
        temp[i] = phi[i]

#storing data phi (output)
np.savetxt('data_1D_training_label.csv', phi, fmt = '%.8f', delimiter=',')

#storing data rho (input)
np.savetxt('data_1D_training_input.csv', data_rho, fmt = '%.8f', delimiter=',')
print (phi)

#plotting hasil
plt.plot(j, phi)
plt.show()
```
